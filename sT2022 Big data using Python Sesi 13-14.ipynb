```python
#define class dog
class Dog:
    def __init__(self,name, age):
        self.name = name
        self.age = age
        
    def sleep(self, x):
        return "zzzz" + " " + x
    def suara(self):
        print("Mengonggong")
    
    def get_name(self):
        return self.name
    def get_age(self):
        return self.age
        
#create an object
d = Dog("Boby", 14)
print(d.get_age())

d2 = Dog("Bill", 10)
print(d2.get_age())
```

```python
#define class dog
class Dog:
    def __init__(self,name, age):
        self.name = name
        self.age = age
        
    def get_name(self):
        return self.name
    def get_age(self):
        return self.age
    def set_age(self,age):
        self.age = age
        
#create an object
d = Dog("Boby", 14)
print(d.get_age())
d.set_age(13)
print(d.get_age())
```

Exercise, buat dua kelas, **Student** (name age, grade, fungsi get_grade) dan **Course**
(nama, max_students, fungsi add_student, average_garde)

```python
class Student:
    def __init__(self, name, age, grade):
        self.name = name
        self.age = age
        self.grade = grade
        
    def get_grade(self):
        return self.grade
    
    def get_name(self):
        return self.name
    
class Course:
    def __init__(self, name, max_students):
        self.name = name
        self.max_students = max_students
        self.students = []
    
    def add_student(self, std):
        if len(self.students) < self.max_students:
            self.students.append(std)
            return True
        return print("Kuota habis untuk ",std.get_name())
    
    def get_average_grade(self):
        val = 0
        for std in self.students:
            val += std.get_grade()
        return val/ len(self.students)
    
s1 = Student("Andi", 18, 96)
s2 = Student("Budi", 18, 70)
s3 = Student("Anto", 18, 50)

MK = Course("Sejarah", 2)
MK.add_student(s1)
MK.add_student(s2)
MK.add_student(s3)
print(MK.students[1].name)
print("Average grade:", MK.get_average_grade())
```

```python
class Person:
    number_of_people = 0
    
    def __init__(self, name):
        self.name  = name
        Person.add_person()
        
    @classmethod
    def number_of_people_(cls):
        return cls.number_of_people
    
    @classmethod
    def add_person(cls):
        cls.number_of_people +=1
        
p1 = Person("Budi")
p2 = Person("Andi")

print(Person.number_of_people_())
```

```python
class Math:
    
    @staticmethod
    def add6(x):
        return x + 6
    
    @staticmethod
    def add10(x):
        return x + 10
    
print(Math.add10(5))
```
